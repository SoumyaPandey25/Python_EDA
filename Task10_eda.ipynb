```python
scores = ['GRADE', 'Cumulative grade point average in the last semester (/4.00)', 'Expected Cumulative grade point average in the graduation (/4.00)']

for col in scores:
    plt.hist(df[col], bins=20)
    plt.title(col)
    plt.xlabel("Score")
    plt.ylabel("Frequency")
    plt.show()
```

```python
for col in scores:
    plt.boxplot(df[col])
    plt.title(col)
    plt.ylabel("Score")
    plt.show()
```

```python
Q1 = df['GRADE'].quantile(0.25)
Q3 = df['GRADE'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_math = df[(df['GRADE'] < lower) | (df['GRADE'] > upper)]
outliers_math.shape
```

```python
col = 'Cumulative grade point average in the last semester (/4.00)' # Corrected column name

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['cgpa_outlier'] = ((df[col] < lower) | (df[col] > upper)).astype(int)
```

```python
df['cgpa_outlier'] = ((df['GRADE'] < lower) |
                      (df['GRADE'] > upper)).astype(int)


#Outliers were detected using the IQR method.
#However, outliers were not removed because low or high scores represent genuine student performance and not data errors.
#Removing them would bias the analysis.
```
